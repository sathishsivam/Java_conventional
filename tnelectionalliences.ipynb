```
votes_individual
```

```
print("Election Prediction 2019")
```

```
import matplotlib.pyplot as plt
```

```
votes_individual.plot()
plt.xlabel("Party (Indexed)")
plt.ylabel("Votes")
plt.title("Individual Votes by the parties")
```
