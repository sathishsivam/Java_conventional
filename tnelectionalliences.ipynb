```
import pandas as pd
```

```
election_result16=pd.read_html("https://en.wikipedia.org/wiki/2016_Tamil_Nadu_Legislative_Assembly_election",header=0)[15]
```

```
election_result16
```

```
election_res16=election_result16.dropna()
```

```
election_res16
```

```
# Current ADMK +PMK (As they stood as a individual parties in Assembly election 2016)
```

```
votes_individual=election_res16[['PARTY','VOTES']]
```

```
votes_individual
```

```
election_res16['VOTES'].sum()
```

```
print("Election Prediction 2019")
```

```
import matplotlib.pyplot as plt
```

```
votes_individual.plot()
plt.xlabel("Party (Indexed)")
plt.ylabel("Votes")
plt.title("Individual Votes by the parties")
```

```
votes_individual.sort_values("VOTES",ascending=False)
```

```
print("Top ten Partys by Votes ")
```

```
top_ten=votes_individual.sort_values("VOTES",ascending=False).head(10)
```
